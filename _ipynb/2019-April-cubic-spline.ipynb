```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
```

```python
waypoints = np.random.randn(3, 2)
times = np.arange(3)

plt.scatter(times, waypoints[:, 0], label='dof0')
plt.scatter(times, waypoints[:, 1], label='dof1')
plt.legend()
plt.show()
```

### CubicSpline interpolation

There are multiple way to select the boundary conditions of a cubic spline interpolation.
One can either let it be 'natural', 'clamped' or fixed a given first-order derivative.
In all cases, CubicSpline is quite natural to work with.

```python
sp = CubicSpline(times, waypoints)
sp2 = CubicSpline(times, waypoints, bc_type='clamped')
sp3 = CubicSpline(times, waypoints, bc_type=((1, [1, 1]), (1, [0.s, 2])))
```

```python
plt.plot(times, sp(times, 1)[:, 0], label="vanilla")
plt.plot(times, sp2(times, 1)[:, 0], label="clamped")
plt.plot(times, sp3(times, 1)[:, 0], label="fixed")
plt.legend()
plt.title("First derivative of splines with different bc")
plt.show()
```
